<b>Importing Libraries</b>

```python
#numpy is used to perform a wide variety of operations on array.
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler,LabelEncoder
```

 <b>Importing Melbourne Housing Market dataset</b>

```python
df=pd.read_csv('melb_data.csv') #reading the csv file
```

```python
#head will display the first 5 elements of the dataset
df.head(5)
```

<b>Data Preprocessing </b>

```python
df.info() #info function is used to give the summary of the dataset
```

```python
df.dtypes #it returns the datatypes of the columns in the data frames
```

```python
df.shape #returns the dimentions of the data frame
```

```python
df.describe() #it returns the description of the data in the dataframe
```

```python
df.isna().sum() #isnull and isna are same functions.They both return the number of missing values
```

```python
for col in df.columns:
    if(df[col].isnull().any() and df[col].dtype!=object):
        df[col].fillna(value=df[col].mean(),inplace=True) #fillna is a function used to replace null values in dataframe
    if(df[col].isnull().any() and df[col].dtype==object):
        df[col].fillna(value=df[col].mode()[0],inplace=True)
```

```python
df.isna().sum() #it returns the number of null values in a particular column.
```

#### Normalizing the data

```python
#in this we are normalizing the data that is scaling the data in the range 0 to 1.
scale=MinMaxScaler()
for col in df.columns:
    if(df[col].dtype!=object):
        df[col]=scale.fit_transform(np.array(df[col]).reshape(-1,1)) #reshape is used to increase 1 dimension of the matrix
df.head(10)
```

<b>Converting categorical variables into quantitative variables </b>

```python
#returns object containing counts of unique values
print(df['Method'].value_counts())
print(df['SellerG'].value_counts())
print(df['Type'].value_counts())
print(df['Regionname'].value_counts())
print(df['CouncilArea'].value_counts())
```

```python
#Label Encoding refers to converting the labels into a numeric form so as to convert them into the machine-readable form.
le=LabelEncoder()
cat_cols=['Type','Method','SellerG','Regionname','CouncilArea']
for col in cat_cols:
    df[col]=le.fit_transform(df[col])

df.head()
```
