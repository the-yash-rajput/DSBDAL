```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df["age"].fillna(int(df.age.mean()),inplace=True) #fillna is used to fill all the null values with the null value.
```

```python
sns.boxplot(x='sex',y='age',hue='survived',data=df) #box plot is used to show outliers in the dataset
```

```python
print("Median Age of survived males",df[(df['survived']==1) & (df['sex']=='male')]["age"].median())
print("Median Age of not survived males",df[(df['survived']==0) & (df['sex']=='male')]["age"].median())
print("Median Age of survived females",df[(df['survived']==1) & (df['sex']=='female')]["age"].median())
print("Median Age of not survived females",df[(df['survived']==0) & (df['sex']=='female')]["age"].median())
```

```python
print("Max. Age of survived male",df[(df['survived']==1) & (df['sex']=='male')]["age"].max())
print("Max. Age of not survived male",df[(df['survived']==0) & (df['sex']=='male')]["age"].max())
print("Max. Age of survived female",df[(df['survived']==1) & (df['sex']=='female')]["age"].max())
print("Max. Age of not survived female",df[(df['survived']==0) & (df['sex']=='female')]["age"].max())
```

##### Here we can see the median age of survived males is approx. 29 and maximum age male survived is approx. 80. While the medain age of males who are not alive is approx. 29 and maximum age male who is not alive is approx. 74

##### Similarly, we can see the median age of survived females is approx. 29 and maximum age female survived is approx. 63. While the medain age of females who are not alive is approx. 29 and maximum age female who is not alive is approx. 57
